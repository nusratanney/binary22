```python
import numpy as np
import matplotlib.pyplot as plt

x = range(1, 16)
y1 = np.random.rand(15)
y2 = np.random.rand(15)

plt.plot(x, y1, marker='o', linestyle='-', label='Vector 1', color='b')
plt.plot(x, y2, marker='s', linestyle='--', label='Vector 2', color='r')

plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Plot of Two Random Vectors')
plt.legend()

plt.show()
```

```python
import seaborn as sns


matrix = np.random.rand(4, 4)
print(matrix)

labels = ['A', 'B', 'C', 'D']

sns.heatmap(matrix, annot=True, cmap='coolwarm', xticklabels=labels, yticklabels=labels)


plt.title("4x4 Random Matrix Heatmap")


plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


A = np.random.randint(0, 10, size=(4, 4))
B = np.random.randint(0, 10, size=(4, 4))

print("A:\n", A)
print("B:\n", B)


matrix_sum = A + B
matrix_diff = A - B
matrix_prod = np.dot(A, B)


indices = range(1, 17)


fig, axes = plt.subplots(1, 3, figsize=(15, 5))


axes[0].bar(indices, matrix_sum.flatten(), color='b')
axes[0].set_title('Matrix Addition')
axes[0].set_xlabel('Index')
axes[0].set_ylabel('Value')


axes[1].bar(indices, matrix_diff.flatten(), color='g')
axes[1].set_title('Matrix Subtraction')
axes[1].set_xlabel('Index')


axes[2].bar(indices, matrix_prod.flatten(), color='r')
axes[2].set_title('Matrix Multiplication')
axes[2].set_xlabel('Index')

plt.tight_layout()
plt.show()
```

**The first set (blue) adds the matrices together, resulting in a range of values.  The second set (green) subtracts them, leading to both positive and negative results.  The third set (red) multiplies them, drastically increasing the values and showing a much wider spread.  Essentially, it visually demonstrates how addition, subtraction, and multiplication affect matrix data.**

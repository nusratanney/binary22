1.Calculate the t-statistic and p-value.

```python
import numpy as np
from scipy import stats

# Given data
data = [5.2, 5.4, 5.1, 5.6, 5.3, 5.8]

# Hypothesized mean
hypothesized_mean = 5.5

# Perform a one-sample t-test
t_stat, p_value = stats.ttest_1samp(data, hypothesized_mean)

# Output results
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")
```

2.Interpret and visualize the result at a 0.05 significance level.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Given data
data = np.array([5.2, 5.4, 5.1, 5.6, 5.3, 5.8])

# Hypothesized mean
mu_0 = 5.5

# Perform one-sample t-test
t_stat, p_value = stats.ttest_1samp(data, mu_0)

# Interpretation
alpha = 0.05  # Significance level
if p_value < alpha:
    result = "Reject H₀: The mean is significantly different from 5.5"
else:
    result = "Fail to reject H₀: No significant difference from 5.5"


plt.figure(figsize=(8,5))
sns.histplot(data, bins=5, kde=True, color="skyblue", edgecolor="black", alpha=0.7)

# Add hypothesized mean
plt.axvline(mu_0, color='red', linestyle='dashed', linewidth=2, label=f"Hypothesized Mean (5.5)")
plt.axvline(np.mean(data), color='green', linestyle='solid', linewidth=2, label=f"Sample Mean ({np.mean(data):.2f})")

# Labels and title
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Distribution of Sample Data")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)

# Display results
plt.show()

# Print test results
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")
print(result)
```
